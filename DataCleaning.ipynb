```python
#importing python modules/libraries
import pandas as pd
import numpy as nump
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
def cleaning(df):  
    print("Here we check how many NAN we have:\n",df.isnull().sum())  #counting how many NAN we have in each attribute
    print("The total number of each attribut are:\n",df.count())  #counting total number of each attribute
    df.dropna(inplace=True)  #removing NAN value rows from dataframe and again checking number of NAN rows in dataframe
    print("Here we'll see the number of NAN we have (should be 0 for each attribute):\n",df.isnull().sum())  #Here we'll see the number of NAN we have (should be 0 for each attribute)
    print("Now we'll see the total number of each attribute after we deleted the NAN rows:\n",df.count())  #Now we'll see the total number of each attribute after we deleted the NAN rows
    df.info()   #printing information of dataframe-should be non-null for each attribute
    print(df.age.unique())    #checking Age ranges
    df.describe()    #table of special calculations
    print(df['country'].unique())   #checking country names
    print("Total Countries: ",len(df['country'].unique()))
    print("Total number of population of world and suicides per year\n",df.groupby(['year']).sum())  #total number of population of world and suicides per year
    return df
    
```

```python
def handeling_data():
    df=pd.read_csv(r"C:\Users\aviv\Downloads\who_suicide_statistics.csv")
    df=cleaning(df)
    df.to_excel('Suicide_Statistics.xlsx',engine='xlsxwriter')
    return df

df=handeling_data()
```
